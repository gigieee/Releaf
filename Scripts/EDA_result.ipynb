```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import os

# ===== PATHS =====
BASE_PROC = "/Users/thanwaratkeratipasuwat/Desktop/dsi314/Processed data"
FILE_MISSING_IN  = os.path.join(BASE_PROC, "bkf_missing_wfo_step4.csv")
FILE_MISSING_OUT = os.path.join(BASE_PROC, "bkf_missing_wfo_step4_ready.csv")

# ===== helpers =====
def nonempty(s: pd.Series) -> pd.Series:
    return s.astype(str).str.strip().ne("") & s.notna()

# ===== load =====
df_missing = pd.read_csv(FILE_MISSING_IN)

# ‡πÉ‡∏´‡πâ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏õ‡∏•‡∏≤‡∏¢‡∏ó‡∏≤‡∏á
for col in ["species_scientific_name", "wfo_full_name", "POWO_name"]:
    if col not in df_missing.columns:
        df_missing[col] = np.nan

# ‡∏≠‡∏±‡∏õ‡πÄ‡∏î‡∏ï species_scientific_name ‡∏ï‡∏≤‡∏°‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç: POWO_name > wfo_full_name > (‡∏Ñ‡πà‡∏≤‡∏î‡∏±‡πâ‡∏á‡πÄ‡∏î‡∏¥‡∏°)
mask_powo = nonempty(df_missing["POWO_name"])
mask_wfo  = nonempty(df_missing["wfo_full_name"])

# ‡∏ô‡∏±‡∏ö‡∏Å‡πà‡∏≠‡∏ô‡πÅ‡∏Å‡πâ
before_nonempty = int(nonempty(df_missing["species_scientific_name"]).sum())

# ‡πÉ‡∏™‡πà POWO_name ‡∏Å‡πà‡∏≠‡∏ô (‡∏ó‡∏±‡∏ö‡πÑ‡∏î‡πâ)
df_missing.loc[mask_powo, "species_scientific_name"] = df_missing.loc[mask_powo, "POWO_name"].astype(str)

# ‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏•‡∏±‡∏á‡πÉ‡∏ä‡πâ POWO_name ‡∏Ñ‡πà‡∏≠‡∏¢‡∏•‡∏á wfo_full_name
mask_still_empty = ~nonempty(df_missing["species_scientific_name"])
df_missing.loc[mask_still_empty & mask_wfo, "species_scientific_name"] = df_missing.loc[mask_still_empty & mask_wfo, "wfo_full_name"].astype(str)

after_nonempty = int(nonempty(df_missing["species_scientific_name"]).sum())

# ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ï‡∏≤‡∏°‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£
keep_cols = [
    "volume", "family_name", "genus_name", "genus_label", "genus_index_parsed",
    "species_scientific_name", "thailand", "distribution", "ecology",
    "family_url", "genus_url", "species_url", "specific_name"
]
# ‡∏ö‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏≠‡∏≤‡∏à‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå‚Äî‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤‡∏ß‡πà‡∏≤‡∏á‡πÉ‡∏´‡πâ
for c in keep_cols:
    if c not in df_missing.columns:
        df_missing[c] = np.nan

df_missing_final = df_missing[keep_cols].copy()

# ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏ú‡∏•
df_missing_final.to_csv(FILE_MISSING_OUT, index=False)
print(f"‚úÖ Saved ‚Üí {FILE_MISSING_OUT}")
print(f"üè∑Ô∏è species_scientific_name non-empty: {before_nonempty} ‚Üí {after_nonempty}")
```

```python
# -*- coding: utf-8 -*-
import pandas as pd
import os

# ===== PATH CONFIG =====
BASE_PROC = "/Users/thanwaratkeratipasuwat/Desktop/dsi314/Processed data"
BASE_RAW  = "/Users/thanwaratkeratipasuwat/Desktop/dsi314/Raw data"

FILE_PRESENT = os.path.join(BASE_PROC, "bkf_present_wfo_ready.csv")
FILE_MISSING = os.path.join(BASE_PROC, "bkf_missing_wfo_step4_ready.csv")
OUT_FILE     = os.path.join(BASE_RAW,  "concat_clean.csv")

# ===== LOAD FILES =====
df_present = pd.read_csv(FILE_PRESENT)
df_missing = pd.read_csv(FILE_MISSING)

print(f"üìò Present rows: {len(df_present)}, columns: {len(df_present.columns)}")
print(f"üìó Missing rows: {len(df_missing)}, columns: {len(df_missing.columns)}")

# ===== ALIGN COLUMNS =====
# (‡∏Å‡∏£‡∏ì‡∏µ‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÑ‡∏°‡πà‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ô ‡πÉ‡∏´‡πâ‡∏£‡∏ß‡∏°‡∏ä‡∏∑‡πà‡∏≠‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏ï‡∏¥‡∏° NaN ‡∏ä‡πà‡∏≠‡∏á‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ)
all_cols = sorted(set(df_present.columns) | set(df_missing.columns))
df_present = df_present.reindex(columns=all_cols)
df_missing = df_missing.reindex(columns=all_cols)

# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏£‡∏∞‡∏ö‡∏∏‡πÅ‡∏´‡∏•‡πà‡∏á‡∏ó‡∏µ‡πà‡∏°‡∏≤
df_present["source"] = "present"
df_missing["source"] = "missing"

# ===== CONCAT =====
df_all = pd.concat([df_present, df_missing], ignore_index=True)

# ===== CLEAN UP (strip text) =====
for c in df_all.select_dtypes(include="object").columns:
    df_all[c] = df_all[c].astype(str).str.strip().replace({"nan": "", "None": ""})

# ===== SAVE =====
df_all.to_csv(OUT_FILE, index=False)
print(f"‚úÖ Saved combined dataset ‚Üí {OUT_FILE}")
print(f"üìä Total rows: {len(df_all)}")
#print(f"üÜî Non-empty wfo_id: {(df_all['wfo_id'].astype(str).str.strip() != '').sum()} / {len(df_all)}")
```

```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import os
import re

# ============ PATHS ============
BASE_RAW  = "/Users/thanwaratkeratipasuwat/Desktop/dsi314/Raw data"
BASE_PROC = "/Users/thanwaratkeratipasuwat/Desktop/dsi314/Processed data"

FILE_ALL     = os.path.join(BASE_RAW,  "concat_clean.csv")
FILE_ECOCROP = os.path.join(BASE_RAW,  "ecocrop_ecology_datasheet.csv")
OUT_MERGED   = os.path.join(BASE_RAW,  "species_factors_merged.csv")
OUT_UNMATCH  = os.path.join(BASE_RAW,  "species_unmatched_for_review.csv")

COL_SPECIES_FACTORS = "Species"   # ‡πÅ‡∏Å‡πâ‡πÉ‡∏´‡πâ‡∏ï‡∏£‡∏á‡πÑ‡∏ü‡∏•‡πå‡∏õ‡∏±‡∏à‡∏à‡∏±‡∏¢‡∏à‡∏£‡∏¥‡∏á‡∏ñ‡πâ‡∏≤‡∏ä‡∏∑‡πà‡∏≠‡∏ï‡πà‡∏≤‡∏á

# ============ HELPERS ============
_rank_tokens = r"(subsp\.|ssp\.|var\.|f\.|forma|cf\.|aff\.|group|agg\.)"

def strip_authors_and_ranks(name: str) -> str:
    if not isinstance(name, str): return ""
    s = name.strip()
    s = re.sub(r"^\s*[√óx]\s*", "", s)
    toks = re.split(r"\s+", s)
    if len(toks) >= 2:
        genus, epithet = toks[0], toks[1]
        if re.fullmatch(_rank_tokens, epithet.lower()):
            epithet = toks[2] if len(toks) >= 3 else ""
        s2 = f"{genus} {epithet}"
    else:
        s2 = s
    s2 = re.sub(_rank_tokens, "", s2, flags=re.IGNORECASE).strip()
    s2 = re.sub(r"[^a-zA-Z0-9\s\-]", " ", s2)
    s2 = re.sub(r"\s+", " ", s2).strip()
    return s2

def canon_key(series: pd.Series) -> pd.Series:
    return (series.astype(str)
                  .str.strip()
                  .str.lower()
                  .str.replace(r"\s+", " ", regex=True))

def genus_of(s: str) -> str:
    return s.split()[0] if isinstance(s, str) and s.strip() else ""

def epithet_of_core(s: str) -> str:
    toks = s.split()
    return toks[1] if len(toks) >= 2 else ""

def nonempty(s: pd.Series) -> pd.Series:
    return s.astype(str).str.strip().ne("") & s.notna()

def unify(df: pd.DataFrame, base: str) -> pd.DataFrame:
    """
    ‡∏£‡∏ß‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏≠‡∏≤‡∏à‡∏°‡∏µ suffix (_fac/_all) ‡πÉ‡∏´‡πâ‡∏°‡∏≤‡∏≠‡∏¢‡∏π‡πà‡∏ä‡∏∑‡πà‡∏≠ base
    ‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç: _fac > _all > base
    """
    for src in (f"{base}_fac", f"{base}_all", base):
        if src in df.columns:
            df[base] = df[src]
            break
    return df

def drop_suffix_cols(df: pd.DataFrame, bases: list) -> pd.DataFrame:
    """‡∏•‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå *_fac / *_all ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏µ‡∏¢‡πå‡∏ó‡∏µ‡πà‡∏£‡∏ß‡∏°‡πÅ‡∏•‡πâ‡∏ß ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Å‡∏±‡∏ô‡∏ä‡∏ô‡πÉ‡∏ô‡∏Å‡∏≤‡∏£ merge ‡∏£‡∏≠‡∏ö‡∏ñ‡∏±‡∏î‡πÑ‡∏õ"""
    to_drop = []
    for b in bases:
        for suf in ("_fac", "_all"):
            c = f"{b}{suf}"
            if c in df.columns:
                to_drop.append(c)
    if to_drop:
        df = df.drop(columns=to_drop)
    return df

# ============ LOAD ============
df_all = pd.read_csv(FILE_ALL)
df_fac = pd.read_csv(FILE_ECOCROP)

if "species_scientific_name" not in df_all.columns:
    raise ValueError("concat_clean.csv ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå species_scientific_name")
if COL_SPECIES_FACTORS not in df_fac.columns:
    raise ValueError(f"‡πÑ‡∏ü‡∏•‡πå‡∏õ‡∏±‡∏à‡∏à‡∏±‡∏¢‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå '{COL_SPECIES_FACTORS}'")

# taxonomy cols ‡∏ó‡∏µ‡πà‡∏à‡∏∞‡πÄ‡∏Å‡πá‡∏ö‡∏à‡∏≤‡∏Å df_all
TAXO_KEEP = [
    "volume", "family_name", "genus_name", "genus_label", "genus_index_parsed",
    "species_scientific_name", "thailand", "distribution", "ecology",
    "family_url", "genus_url", "species_url", "specific_name", "wfo_id", "source"
]
for c in TAXO_KEEP:
    if c not in df_all.columns:
        df_all[c] = np.nan

# ============ BUILD KEYS ============
df_all["__name_full__"] = canon_key(df_all["species_scientific_name"])
df_all["__name_core__"] = canon_key(df_all["species_scientific_name"].map(strip_authors_and_ranks))
df_all["__genus_l__"]   = df_all["species_scientific_name"].astype(str).map(genus_of).str.lower()
df_all["__epithet__"]   = df_all["species_scientific_name"].map(strip_authors_and_ranks).map(epithet_of_core).str.lower()

df_fac["__name_full__"] = canon_key(df_fac[COL_SPECIES_FACTORS])
df_fac["__name_core__"] = canon_key(df_fac[COL_SPECIES_FACTORS].map(strip_authors_and_ranks))
df_fac["__genus_l__"]   = df_fac[COL_SPECIES_FACTORS].astype(str).map(genus_of).str.lower()
df_fac["__epithet__"]   = df_fac[COL_SPECIES_FACTORS].map(strip_authors_and_ranks).map(epithet_of_core).str.lower()

# ‡∏Å‡∏±‡∏ô‡∏Ñ‡∏µ‡∏¢‡πå‡∏ã‡πâ‡∏≥‡∏ù‡∏±‡πà‡∏á‡∏õ‡∏±‡∏à‡∏à‡∏±‡∏¢
fac_full = df_fac.dropna(subset=["__name_full__"]).drop_duplicates("__name_full__", keep="first")
fac_core = df_fac.dropna(subset=["__name_core__"]).drop_duplicates("__name_core__", keep="first")

# ============ PASS A: full exact ============
leftA  = fac_full[[COL_SPECIES_FACTORS, "__name_full__", "__name_core__", "__genus_l__", "__epithet__"]].copy()
rightA = df_all[TAXO_KEEP + ["__name_full__", "__name_core__", "__genus_l__", "__epithet__"]].copy()

mA = leftA.merge(rightA, on="__name_full__", how="left", suffixes=("_fac", "_all"))
for col in ["__name_full__", "__name_core__", "__genus_l__", "__epithet__"]:
    mA = unify(mA, col)
mA = drop_suffix_cols(mA, ["__name_full__", "__name_core__", "__genus_l__", "__epithet__"])

mA["match_type"] = np.where(mA["species_scientific_name"].notna(), "A_full_exact", None)

# ============ PASS B: core exact ============
unmatchedA = mA[mA["species_scientific_name"].isna()].copy()
leftB  = unmatchedA[[COL_SPECIES_FACTORS, "__name_core__", "__genus_l__", "__epithet__", "__name_full__"]].copy()
rightB = df_all[TAXO_KEEP + ["__name_core__", "__genus_l__", "__epithet__", "__name_full__"]].copy()

mB = leftB.merge(rightB, on="__name_core__", how="left", suffixes=("_fac", "_all"))
for col in ["__name_full__", "__name_core__", "__genus_l__", "__epithet__"]:
    mB = unify(mB, col)
mB = drop_suffix_cols(mB, ["__name_full__", "__name_core__", "__genus_l__", "__epithet__"])

mB["match_type"] = np.where(mB["species_scientific_name"].notna(), "B_core_exact", None)

# ============ PASS C: same genus + epithet exact ============
unmatchedB = mB[mB["species_scientific_name"].isna()].copy()

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì epithet ‡∏ù‡∏±‡πà‡∏á‡∏õ‡∏±‡∏à‡∏à‡∏±‡∏¢ (‡∏à‡∏≤‡∏Å __name_core__) ‡πÉ‡∏´‡∏°‡πà‡∏ö‡∏ô leftC ‡∏ó‡∏µ‡πà‡∏™‡∏∞‡∏≠‡∏≤‡∏î
leftC = unmatchedB[[COL_SPECIES_FACTORS, "__genus_l__", "__name_core__"]].copy()
leftC["__epithet__fac"] = leftC["__name_core__"].astype(str).map(epithet_of_core)

rightC = df_all[TAXO_KEEP + ["__genus_l__", "__epithet__"]].copy()

mC = leftC.merge(rightC, on="__genus_l__", how="left", suffixes=("_fac", "_all"))
# unify ‡πÄ‡∏â‡∏û‡∏≤‡∏∞ key ‡∏ó‡∏µ‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô
mC = unify(mC, "__genus_l__")
# ‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö epithet
mC = mC[mC["__epithet__fac"].fillna("") == mC["__epithet__"].fillna("")]
mC["match_type"] = np.where(mC["species_scientific_name"].notna(), "C_genus_epithet", None)

# ============ PACK ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ä‡∏∏‡∏î‡πÄ‡∏î‡∏µ‡∏¢‡∏ß ============
def pack(df, extra_cols=None):
    cols = [COL_SPECIES_FACTORS, "__name_full__", "__name_core__", "__genus_l__", "__epithet__", "match_type"] + TAXO_KEEP
    if extra_cols:
        cols += extra_cols
    keep = [c for c in cols if c in df.columns]
    return df[keep].copy()

gotA = pack(mA[mA["species_scientific_name"].notna()])
gotB = pack(mB[mB["species_scientific_name"].notna()])
# ‡πÉ‡∏ô mC ‡πÑ‡∏°‡πà‡∏°‡∏µ __name_full__/__name_core__ ‡∏ù‡∏±‡πà‡∏á right ‡πÉ‡∏´‡πâ‡∏Ñ‡∏á‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏ó‡∏µ‡πà‡∏°‡∏µ
gotC = pack(mC[mC["species_scientific_name"].notna()], extra_cols=["__epithet__fac"])

bridge = pd.concat([gotA, gotB, gotC], ignore_index=True) \
           .drop_duplicates(subset=[COL_SPECIES_FACTORS], keep="first")

# ============ unmatched ============
unmatched_keys = set(df_fac[COL_SPECIES_FACTORS]) - set(bridge[COL_SPECIES_FACTORS])
unmatched = df_fac[df_fac[COL_SPECIES_FACTORS].isin(unmatched_keys)].copy()
unmatched["match_type"] = "Z_unmatched"

# ============ ‡πÅ‡∏ô‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏õ‡∏±‡∏à‡∏à‡∏±‡∏¢‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î ============
factor_cols = [c for c in df_fac.columns if c not in {
    COL_SPECIES_FACTORS, "__name_full__", "__name_core__", "__genus_l__", "__epithet__"
}]
bridge_full = bridge.merge(
    df_fac[[COL_SPECIES_FACTORS] + factor_cols].drop_duplicates(COL_SPECIES_FACTORS, keep="first"),
    on=COL_SPECIES_FACTORS, how="left"
)

# ========= SAVE =========
out_cols = TAXO_KEEP + ["match_type", COL_SPECIES_FACTORS] + factor_cols
out_cols = [c for c in out_cols if c in bridge_full.columns]
bridge_full[out_cols].to_csv(OUT_MERGED, index=False)

unmatched[[COL_SPECIES_FACTORS] + factor_cols + ["__name_full__", "__name_core__", "__genus_l__", "__epithet__", "match_type"]] \
    .to_csv(OUT_UNMATCH, index=False)

print(f"‚úÖ Saved merged factors ‚Üí {OUT_MERGED}")
print(f"üü° Unmatched species (from factors): {len(unmatched)} ‚Üí {OUT_UNMATCH}")
if "match_type" in bridge.columns:
    print("üìä match_type counts:")
    print(bridge["match_type"].value_counts())
```

```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import os, re
from collections import defaultdict

# ========== PATHS ==========
BASE_RAW  = "/Users/thanwaratkeratipasuwat/Desktop/dsi314/Raw data"
BASE_PROC = "/Users/thanwaratkeratipasuwat/Desktop/dsi314/Processed data"

FILE_ALL        = os.path.join(BASE_RAW,  "concat_clean.csv")
FILE_ECOCROP    = os.path.join(BASE_RAW,  "ecocrop_ecology_datasheet.csv")
FILE_UNMATCHED  = os.path.join(BASE_RAW,  "species_unmatched_for_review.csv")  # ‡∏à‡∏≤‡∏Å‡∏£‡∏≠‡∏ö‡∏Å‡πà‡∏≠‡∏ô
OUT_FUZZY       = os.path.join(BASE_RAW,  "species_fuzzy_matches.csv")
OUT_MERGED_V2   = os.path.join(BASE_RAW,  "species_factors_merged_v2.csv")
OUT_UNMATCH_V2  = os.path.join(BASE_RAW,  "species_unmatched_after_fuzzy.csv")

COL_SPECIES_FACTORS = "Species"   # ‡∏õ‡∏£‡∏±‡∏ö‡∏ñ‡πâ‡∏≤‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ï‡πà‡∏≤‡∏á‡πÑ‡∏õ
FUZZY_THRESHOLD = 94                   # ‡∏õ‡∏£‡∏±‡∏ö‡πÑ‡∏î‡πâ 92‚Äì96 ‡∏ï‡∏≤‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏ó‡∏µ‡πà‡∏¢‡∏≠‡∏°‡∏£‡∏±‡∏ö

# ========== DEPS ==========
try:
    from rapidfuzz import fuzz, process
except Exception as e:
    raise SystemExit("‡∏ï‡πâ‡∏≠‡∏á‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á rapidfuzz ‡∏Å‡πà‡∏≠‡∏ô‡∏ô‡∏∞‡∏Ñ‡∏∞: pip install rapidfuzz")

# ========== HELPERS ==========
_rank_tokens = r"(subsp\.|ssp\.|var\.|f\.|forma|cf\.|aff\.|group|agg\.)"
_stop_tokens = {"cf", "aff", "sp", "spp", "species", "indet"}

def strip_authors_and_ranks(name: str) -> str:
    if not isinstance(name, str): return ""
    s = name.strip()
    s = re.sub(r"^\s*[√óx]\s*", "", s)  # ‡∏•‡∏ö‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢ hybrid ‡∏ó‡∏µ‡πà‡∏´‡∏±‡∏ß
    toks = re.split(r"\s+", s)
    # ‡πÄ‡∏≠‡∏≤‡πÅ‡∏Ñ‡πà‡∏™‡∏≠‡∏á‡∏Ñ‡∏≥‡πÅ‡∏£‡∏Å‡πÄ‡∏õ‡πá‡∏ô‡∏´‡∏•‡∏±‡∏Å (genus + epithet)
    if len(toks) >= 2:
        genus, epithet = toks[0], toks[1]
        if re.fullmatch(_rank_tokens, epithet.lower()):
            epithet = toks[2] if len(toks) >= 3 else ""
        s2 = f"{genus} {epithet}"
    else:
        s2 = s
    # ‡∏•‡∏ö rank tokens ‡πÅ‡∏•‡∏∞‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î
    s2 = re.sub(_rank_tokens, "", s2, flags=re.IGNORECASE).strip()
    s2 = re.sub(r"[^a-zA-Z0-9\s\-]", " ", s2)
    s2 = re.sub(r"\s+", " ", s2).strip()
    # ‡∏•‡∏ö stop tokens ‡∏ó‡∏µ‡πà‡∏õ‡∏•‡∏≤‡∏¢
    parts = s2.split()
    parts = [p for p in parts if p.lower() not in _stop_tokens]
    return " ".join(parts)

def canon(s: pd.Series) -> pd.Series:
    return (s.astype(str).str.strip().str.lower().str.replace(r"\s+", " ", regex=True))

def genus_of(s: str) -> str:
    return s.split()[0] if isinstance(s, str) and s.strip() else ""

def epithet_of_core(s: str) -> str:
    toks = s.split()
    return toks[1] if len(toks) >= 2 else ""

def nonempty(s: pd.Series) -> pd.Series:
    return s.astype(str).str.strip().ne("") & s.notna()

# ========== LOAD ==========
df_all = pd.read_csv(FILE_ALL)
df_fac = pd.read_csv(FILE_ECOCROP)
df_unm = pd.read_csv(FILE_UNMATCHED)

if "species_scientific_name" not in df_all.columns:
    raise ValueError("concat_clean.csv ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå species_scientific_name")
if COL_SPECIES_FACTORS not in df_fac.columns:
    raise ValueError(f"ecocrop ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå '{COL_SPECIES_FACTORS}'")
if COL_SPECIES_FACTORS not in df_unm.columns:
    raise ValueError(f"unmatched ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå '{COL_SPECIES_FACTORS}'")

# taxonomy cols ‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÅ‡∏ô‡∏ö‡∏ï‡∏≠‡∏ô merge
TAXO_KEEP = [
    "volume","family_name","genus_name","genus_label","genus_index_parsed",
    "species_scientific_name","thailand","distribution","ecology",
    "family_url","genus_url","species_url","specific_name","wfo_id","source"
]
for c in TAXO_KEEP:
    if c not in df_all.columns:
        df_all[c] = np.nan

# ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ñ‡∏µ‡∏¢‡πå‡πÅ‡∏ö‡∏ö core + genus
df_all["__core__"]  = canon(df_all["species_scientific_name"].map(strip_authors_and_ranks))
df_all["__genus__"] = canon(df_all["species_scientific_name"].map(genus_of))
df_all["__epithet__"] = canon(df_all["species_scientific_name"].map(strip_authors_and_ranks).map(epithet_of_core))

df_unm["__core__"]  = canon(df_unm[COL_SPECIES_FACTORS].map(strip_authors_and_ranks))
df_unm["__genus__"] = canon(df_unm[COL_SPECIES_FACTORS].map(genus_of))
df_unm["__epithet__"] = canon(df_unm[COL_SPECIES_FACTORS].map(strip_authors_and_ranks).map(epithet_of_core))

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏î‡∏±‡∏ä‡∏ô‡∏µ candidate ‡∏ï‡πà‡∏≠ genus ‡∏à‡∏≤‡∏Å df_all (‡∏•‡∏î search space)
genus_to_names = defaultdict(list)
genus_to_rows  = defaultdict(list)
for idx, row in df_all.iterrows():
    g = row["__genus__"]
    if not g: 
        continue
    genus_to_names[g].append(row["__core__"])
    genus_to_rows[g].append(idx)

# ========== FUZZY MATCH ‡∏†‡∏≤‡∏¢‡πÉ‡∏ô GENUS ==========
matches = []
for idx, r in df_unm.iterrows():
    g = r["__genus__"]
    q = r["__core__"]
    if not g or not q:
        continue
    candidates = genus_to_names.get(g, [])
    if not candidates:
        continue
    # ‡πÉ‡∏ä‡πâ ratio ‡∏ó‡∏µ‡πà robust ‡∏Å‡∏ß‡πà‡∏≤ ‚Äî WRatio ‡∏´‡∏£‡∏∑‡∏≠ token_set_ratio ‡∏Å‡πá‡πÑ‡∏î‡πâ
    # ‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà‡πÉ‡∏ä‡πâ WRatio (‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Å‡∏≤‡∏£‡πÅ‡∏ó‡∏£‡∏Å/‡∏•‡∏ö/‡∏™‡∏•‡∏±‡∏ö‡πÑ‡∏î‡πâ‡∏î‡∏µ)
    scored = process.extract(
        q, candidates, scorer=fuzz.WRatio, score_cutoff=FUZZY_THRESHOLD, limit=5
    )
    if not scored:
        continue
    # tie-breaker: ‡∏ñ‡πâ‡∏≤‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏ó‡πà‡∏≤‡∏Å‡∏±‡∏ô ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏ä‡∏∑‡πà‡∏≠‡∏ó‡∏µ‡πà‡∏¢‡∏≤‡∏ß‡∏Å‡∏ß‡πà‡∏≤ (‡∏°‡∏±‡∏Å‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡∏ó‡∏µ‡πà‡∏Ñ‡∏£‡∏ö‡∏Å‡∏ß‡πà‡∏≤/‡πÉ‡∏Å‡∏•‡πâ accepted)
    best_score = max(s[1] for s in scored)
    best = [s for s in scored if s[1] == best_score]
    # map ‡∏Å‡∏•‡∏±‡∏ö‡πÑ‡∏õ‡∏´‡∏≤ index ‡∏ù‡∏±‡πà‡∏á df_all
    pick = None
    if len(best) == 1:
        pick = best[0]
    else:
        # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏ä‡∏∑‡πà‡∏≠‡∏ó‡∏µ‡πà‡∏¢‡∏≤‡∏ß‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
        best_sorted = sorted(best, key=lambda x: len(x[0]), reverse=True)
        pick = best_sorted[0]

    cand_name = pick[0]
    score     = pick[1]
    # ‡∏´‡∏≤ index ‡∏Ç‡∏≠‡∏á‡∏ä‡∏∑‡πà‡∏≠ candidate ‡∏ô‡∏µ‡πâ‡πÉ‡∏ô df_all (‡∏†‡∏≤‡∏¢‡πÉ‡∏ô genus ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô)
    cand_list = genus_to_names[g]
    pos = [i for i, n in enumerate(cand_list) if n == cand_name]
    if not pos:
        continue
    row_idx = genus_to_rows[g][pos[0]]
    row_all = df_all.iloc[row_idx]

    matches.append({
        "Species 2568": r[COL_SPECIES_FACTORS],
        "query_core": q,
        "query_genus": g,
        "matched_core": cand_name,
        "matched_species_scientific_name": row_all["species_scientific_name"],
        "matched_wfo_id": row_all.get("wfo_id", np.nan),
        "score": score
    })

df_fuzzy = pd.DataFrame(matches).drop_duplicates(subset=["Species 2568"], keep="first")
df_fuzzy.to_csv(OUT_FUZZY, index=False)

print(f"‚úÖ Fuzzy matched {len(df_fuzzy)} rows with threshold ‚â• {FUZZY_THRESHOLD}")
print(f"‚Üí Saved: {OUT_FUZZY}")

# ========== ‡∏£‡∏ß‡∏° fuzzy ‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏õ‡πá‡∏ô merged v2 ==========
# ‡πÅ‡∏ô‡∏ö taxonomy ‡∏à‡∏≤‡∏Å df_all ‡∏ï‡∏≤‡∏° matched_core
df_all_core = df_all[["species_scientific_name","wfo_id","__core__"] + TAXO_KEEP].copy()
df_all_core = df_all_core.drop_duplicates("__core__", keep="first").rename(columns={"__core__": "matched_core"})

df_fuzzy_merge = df_fuzzy.merge(df_all_core, on="matched_core", how="left")

# ‡πÅ‡∏ô‡∏ö factor columns ‡πÄ‡∏ï‡πá‡∏°‡∏à‡∏≤‡∏Å ecocrop
factor_cols = [c for c in df_fac.columns if c != COL_SPECIES_FACTORS]
df_fuzzy_full = df_fuzzy_merge.merge(
    df_fac[[COL_SPECIES_FACTORS] + factor_cols].drop_duplicates(COL_SPECIES_FACTORS, keep="first"),
    left_on="Species 2568", right_on=COL_SPECIES_FACTORS, how="left"
)

# ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å merged V2 (‡πÄ‡∏â‡∏û‡∏≤‡∏∞ fuzzy)
df_fuzzy_full.to_csv(OUT_MERGED_V2, index=False)

# ========== ‡∏™‡∏£‡πâ‡∏≤‡∏á unmatched ‡∏´‡∏•‡∏±‡∏á fuzzy ==========
matched_keys = set(df_fuzzy["Species 2568"])
unmatched_v2 = df_unm[~df_unm[COL_SPECIES_FACTORS].isin(matched_keys)].copy()
unmatched_v2.to_csv(OUT_UNMATCH_V2, index=False)

print(f"üü° Unmatched after fuzzy: {len(unmatched_v2)} ‚Üí {OUT_UNMATCH_V2}")
```
